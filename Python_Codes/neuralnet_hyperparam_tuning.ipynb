```python
import numpy as np
import pandas as pd
```

```python
poi = pd.read_csv("../data/POIs_tracts.csv")
#list(poi.columns.values)
droplist = list(['Unnamed: 0',
 'geometry',
 'statefp10',
 'name10',
 'commarea_n',
 'namelsad10',
 'commarea',
 'notes',
 'countyfp10',
 'centroid'])
poi = poi.drop(droplist, axis=1)
poi.head(1)
len(poi)

# Read poi data
```

```python
poi.head()
```

```python
# rides = pd.read_csv(r'C:\Users\mudd\Documents\F19\CTYPLAN 257\Scripts\initialMondays.csv', sep = ',')
# rides.head(1)
ride = poi[['geoid10']].copy()
s = ride['geoid10'].tolist()
ride['dest'] = np.random.choice(s, len(s))
ride['num'] = np.random.choice(10, len(s))

# Create working dataset for ride
```

```python
dem = pd.read_csv("../data/Pop_data.csv")
print(list(dem.columns.values))
keeplist = ['GEO.id2',
             'Number - Both sexes; Total population',
             'Number - Both sexes; Total population - Median age (years)',
             'Number - Male; Total population - Median age (years)',
             'Number - Female; Total population - Median age (years)']
dem = dem[keeplist]
len(dem)

# Create dem dataframe from pop_data, filter for reqd columns
```

```python
dem = dem.rename(columns={'GEO.id2': 'geo_id2', 
                         'Number - Both sexes; Total population': 'tot_pop',
                         'Number - Both sexes; Total population - Median age (years)': 'tot_pop_dev',
                         'Number - Male; Total population - Median age (years)': 'male_dev',
                         'Number - Female; Total population - Median age (years)': 'female_dev'})

# Rename columns
```

```python
print(poi.columns)
print(dem.columns)

# Column check
```

```python
master = pd.merge(poi,dem,left_on='geoid10',right_on='geo_id2',how='left')
# master.head(1)

# Merge poi with dem 
```
